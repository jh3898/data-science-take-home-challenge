# URL Parsing Challenge

```python
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#df = pd.read_csv('url_list.txt',header= None)
#df.head()
fields = ['hotel.checkin','hotel.customMinimumPriceFilter','hotel.customMaximumPriceFilter','hotel.freeCancellation',
          'hotel.stars_5','hotel.stars_4','hotel.stars_3','hotel.stars_2','hotel.stars_1','hotel.max_score','hotel.min_score',
          'hotel.couponCode','hotel.adults','hotel.city','hotel.children','hotel.amenities','hotel.checkout','hotel.search_page']
#dic = zip(fields,np.zeros(len(fields)))
def clean_data(line1):
    dic = dict.fromkeys(fields)
    for word in line1:
        word = word.strip('+').split('=')
        dic[word[0]]=word[1]
    #dic= pd.Series(dic.values()) #index = dic.keys()
    #line1= dic
    return dic.values()
    
```

```python
line0 =[]
with open('url_list.txt','rt') as f:
    data0 =[line for line in f]
    for line in data0:
        line = line[line.find('?')+1:-1].strip('+').split('&')
        #print(line)
        line =clean_data(line)
        line0.append(line)
        #print(line)
    
df0 =pd.DataFrame(line0,columns = fields)
```

```python
df0['hotel.city'] =df0['hotel.city'].apply(lambda x: x.replace('+', ' '))    #.split(',')
df0['hotel.checkin'] = pd.to_datetime(df0['hotel.checkin'])
df0['hotel.checkout'] = pd.to_datetime(df0['hotel.checkout'])
```
